## I. Energy community [@home+45min]

Increasing the efficiency of energy systems including humans, like micro-grids or buildings, leads to a common issue: they can't be controlled as physical system where human are generally considered as disturbances or, at most, as predictive controllers. Indeed, increasing efficiency increases the sensitivity to the complexity of human behaviour, which depends on complex cognitive processes, high perception capabilities, deliberative behaviors but also on social considerations. For instance, the commitment theory suggests that doing small steps with commitment in front of other people increase the willingness to change. From this idea, the concept of energy community, with its social dimension, has been promoted by Europe and adapted in each European contry. Instead of considering human actors as disturbance or controlled systems, it suggests to develop cooperative and interactive systems where the goals are shared.

## II. Measuring performances of energy systems with humans [15min]

### Self-consumption

To measure the energy locally produced and consumed locally, the self-consumption indicator is used. It is part of the production that is consumed locally: the closer to 100%, the less the grid is used.

![selfconsumption.png](doc1/selfconsumption.png)

Maximizing the self-consumption indicator leads to maximize your profit with more flexibility while keeping the same total energy consumption. It is related to the self consumption model, which encourages households to increase their flexibility. It's also a way for leading people to better understand their consumption, and becoming then more sober. A French incentive to promote individual self-consumption is appearing in the Journal Officiel from 25/02/2017. Let's compare the total sale of produced PV electricity with the self-consumption model where only the overage is sold.

a) In case of total sale, depending on the size and the type of integration of the PV plant, the total PV electricity produced is bought between 15 to 18cts/kWh for smallest plant whereas the electricity is bought at about 17cts/kWh. There is no contribution to the investment.

b) In case of self consumption, the energy produced locally is consumed locally i.e. at no cost. In case of overage, for plants lower than 9kWp, the PV energy produced is bought 10cts/kWh anf 6cts/kWh for bigger plants. Conversely, if the consumption is lower than the production, the regular tariff for the grid applied: about 17cts/kWh. Additionally, a contribution to the investment is done by the state: 0.39€/Wp if the plant ≤3kWp (1170€ for 3kWp), 0.29€/Wp if the plant belongs to 3-9kWp, 0.19€/Wp for 9-36kWp and 0.09€/Wp for 36-100kWp plants.

Obviously, the second model is much more interesting and promoted. Nevertheless, we may wonder that with a physical perspective, both installations are similar, only the way of counting is changing. So why? Once again, the reason is that PV in home is a system with humans. It appears than with the self-consumption model, people analyze their consumption and production to maximize their profits and therefore feel more concerned by their consumption and, as a side effect, might discover energy wastes.

### Self-sufficiency

To measure the energy locally consumed that is produced locally, the self-production, also called self-sufficiency, is used. The highest percentage this indicator, the less the household needs energy from the grid.

![selfproduction.png](doc1/selfproduction.png)

With self-sufficiency indicator, a household may consummed a lot during the PV production period while, with self-consumption, a household may consumed a lot out of the PV production period.

### Dependency

The dependency, also named Net Energy Exchange with the Grid (NEEG), is a way to combine the 2 previous indicateurs: it consists in the difference between consumption and production. Minimizing the dependency amounts to modify consumption and production for a perfect match.

![dependency.png](doc1/dependency.png)



### Measuring the human actor effort (proposal)

Measuring how an energy community member contributes to the community can be done with an effort indicator $E$.

Let $P_i^\text{predict}(t)$ be the predicted consumption of an energy community member with $\rho_i$ tenth, i.e. $\sum_i \rho_i = 1$, contribution to the community and $P_i^\text{actual}(t)$ his actual consumption.

Let $P^\text{ideal}(t)$ be the ideal community consumption perfectly matching the PV production (no dependency with the grid).

The ideal contribution of each community member is given by:

$$
E_i(t) = \left| P_i^\text{actual}(t) - P_i^\text{predict}(t) - \rho_i \left(P^\text{ideal}(t) - \sum_i P_i^\text{predict}(t) \right) \right|
$$
where $\rho_i \left(P^\text{ideal}(t) - \sum_i P_i^\text{predict}(t) \right)$ stands for the ideal contribution the member $i$ should do, and $\left(P_i^\text{actual}(t) - P_i^\text{predict}(t)\right)$ for his actual contribution to reach the perfect matching.

```python
from buildingenergy import openweather, solar
import plotly.graph_objs as go

pv_efficiency = 0.13
pv_surface = 1

site_weather_data = openweather.OpenWeatherMapJsonReader('grenoble_weather2015-2019.json', from_stringdate='16/02/2015 0:00:00', to_stringdate='01/02/2016 0:00:00', sea_level_in_meter=330, albedo=.1).site_weather_data

solar_model = solar.SolarModel(site_weather_data=site_weather_data)
datetimes=solar_model.site_data.get('datetime')

slope = 90
exposure = 0

def pv_production(slope, exposure):
    phisun = solar_model.solar_irradiations(slope_in_deg=slope, exposure_in_deg=exposure)['total']
    return [P * pv_efficiency * pv_surface for P in phisun]

################################ develop your own code here

################################ end of your own code 
production = pv_production(slope, exposure)

fig = go.Figure(go.Scatter(x=datetimes, y=production, name='predicted supplied powers'))
fig.show()
```
